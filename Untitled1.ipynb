```python
# Operadores aritiméticos

# Pedir ao usuário para digitar 02 números

# Entrada

a = int (input("digite um número: "))
b = int (input ("digite um número:"))

# Processamento

# Soma

soma = a + b

subtracao = a - b

multiplicação = a *  b

divisão = a / b

# Saída

print("")
print("----------OPERADORES ARITMÉTICOS------------")
print("")
print("----------SOMA---------")
print("")
print(f"o resultado da soma {a} mais {b} é igual a ; {soma}")
print("")
print("-------------SUBTRAÇÃO-----------")
print("")
print(f"o resultado da DIFERENÇA de {a} MAIS {b} é  igual a: {subtracao}")
print("")
print("-------------MULTIPLICAÇÃO--------------")
print("")
print(f"o resultado da MULTIPLICAÇÃO é {a} * {b} é igual a: {multiplicação}")
print("")
print("------------DIVISÃO-----------")
print("")
print(f" o resultado da DIVISÃO é {a} / {b} é igual a: {divisão}")
print("-------------------------------------")
```
