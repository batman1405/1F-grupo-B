```python
# Escrever um programa que o usuário digita um número e o programa imprimi até o número digitado.

# Pede o número ao usuário e converte para inteiro
numero = int(input("Digite um número: "))

# O range vai de 1 até o número digitado (+1 para incluir o próprio número)
for i in range(1, numero + 1):
    print(i)
```

```python
# Escrever um programa que o usuário digita um número e o programa imprimi o número digitado até o zero (contagem regressiva).

# Pede o número ao usuário
numero = int(input("Digite um número para iniciar a contagem regressiva: "))

# range(início, fim, passo)
# Começa no número, vai até -1 (para incluir o zero) e subtrai 1 a cada volta
for i in range(numero, -1, -1):
    print(i)
```

```python
# Escrever um programa que o usuário digita um número e o programa imprimi os números pares.

# Pede o número ao usuário
numero = int(input("Digite um número: "))

# Começa no 0, vai até o número (+1 para incluir) e pula de 2 em 2
for i in range(0, numero + 1, 2):
    print(i)
```

```python
# Escrever um programa que o usuário digita um número e o programa imprimi os números impares.

numero = int(input("Digite um número: "))

# Começa no 1 (primeiro ímpar), vai até o número e pula de 2 em 2
for i in range(1, numero + 1, 2):
    print(i)
```

```python
# Escrever um programa que o usuário digita um número e programa soma os números até o número digitado.

numero = int(input("Digite um número: "))

soma = 0  # Variável para acumular o total

for i in range(1, numero + 1):
    soma = soma + i  # Adiciona o número atual ao total

print(f"A soma de todos os números de 1 até {numero} é: {soma}")
```

```python
# Escrever um programa que o usuário digita uma palavra e o programa conta as letras da palavra.

palavra = input("Digite uma palavra: ")

# A função len() conta a quantidade de caracteres
quantidade_letras = len(palavra)

print(f"A palavra '{palavra}' tem {quantidade_letras} letras.")
```
